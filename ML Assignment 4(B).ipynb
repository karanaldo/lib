```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import silhouette_score
from sklearn.metrics import pairwise_distances
from sklearn import metrics
from scipy.spatial import distance_matrix
import seaborn as sns
```

```python
df = pd.read_csv("CC GENERAL.csv")
df
```

```python
df.isnull().sum()
```

```python
# Handle any missing values (e.g., fill with zeros)
df = df.fillna(0)
df
```

```python
# Drop non-numeric columns (e.g., customer ID)
df = df.select_dtypes(include=[np.number])
df
```

```python
# Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
# Determine the optimal number of clusters using the Silhouette Method
silhouette_scores = []
k_values = range(2, 10)

for k in k_values:
    kmedoids = KMedoids(n_clusters=k, metric="euclidean", random_state=0)
    labels = kmedoids.fit_predict(df_scaled)
    silhouette_avg = silhouette_score(df_scaled, labels, metric="euclidean")
    silhouette_scores.append(silhouette_avg)

# Plot the Silhouette scores
plt.figure(figsize=(8, 6))
plt.plot(k_values, silhouette_scores, marker='o', linestyle='-', color='b')
plt.title('Silhouette Method for Optimal K (K-Medoids)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

```python
np.unique(labels)
```

```python
# Assuming you have 'kmedoids' as your K-Medoids model
cluster_centers = df_scaled[kmedoids.medoid_indices_]
plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=labels, cmap='rainbow')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='black', marker='x')
plt.title('K-Medoids Clustering with Cluster Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
